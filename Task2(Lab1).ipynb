```python
import pandas as pd
path ="/content/drive/MyDrive/Colab Notebooks/Au_nanoparticle_dataset.csv"
df = pd.read_csv(path)
df.head()
```

```python
df_filtered = df.drop(['N_total', 'N_bulk', 'N_surface', 'R_avg'],axis=1)
df_filtered.head()
```

```python
df_filtered.sample(20)
```

```python
df_selected = df[['N_total', 'N_bulk', 'N_surface','R_avg']]
df_selected.head()
```

```python
mean = df_selected.mean()
print(mean)
```

```python
sd = df_selected.std()
print(sd)
```

```python
quantile = df_selected.quantile([0.25,0.5,0.75])
print(quantile)
```

```python
import matplotlib.pyplot as plt
```

```python
df_selected.hist(layout=(1,4), figsize=(20,5), bins=30, color='skyblue')
plt.show()
```

```python
import seaborn as sns

sns.pairplot(df_selected)
plt.show()
```

```python
g = sns.PairGrid(df_selected)
g.map_upper(sns.histplot)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
```
