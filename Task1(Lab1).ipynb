```python
plt.hist(np.random.exponential(scale=1.0, size=(100000,1)), density=True, bins=100, histtype="step", color="blue", label="Rand_exp")
plt.hist(np.random.normal(loc=0, scale=1, size=100000), density=True, bins=100, histtype="step", color="red", label="Rand_normal")
plt.hist(np.random.uniform(low=0, high=1.0, size=100000), density=True, bins=100, histtype="step", color="green", label="Rand_uniform")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X,Y = np.meshgrid(x,y)

Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)

plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as pyplot
%matplotlib inline
import seaborn as sns

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
df.head()
```

```python
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = df[features]

pearson_corr = data.corr(method="pearson")
spearman_corr = data.corr(method="spearman")

plt.figure(figsize=(8,6))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Pearson Correlation Heatmap")

plt.figure(figsize=(8,6))
sns.heatmap(spearman_corr, annot=True, cmap='cool', fmt=".2f")
plt.title("Spearman Correlation Heatmap")
plt.show()
```
